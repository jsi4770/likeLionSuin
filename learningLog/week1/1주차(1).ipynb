```python
df = pd.read_csv('titanic - titanic.csv')
print(df.head())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

sns.barplot(x='Sex', y='Survived', data=df)
plt.show()
```

```python
survivors = df[df['Survived'] == 1]
```

```python
pclass_survivors = survivors.groupby('Pclass')['Survived'].count().reset_index()

print("--- 등급별 생존자 명수 결과 ---")
print(pclass_survivors)
```

```python
sns.barplot(x='Pclass', y='Survived', data=pclass_survivors, palette='viridis')

plt.title('Number of Survivors by Class')
plt.xlabel('Ticket Class (1st, 2nd, 3rd)')
plt.ylabel('Count of Survivors')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class (Percentage)')
plt.show()
```
